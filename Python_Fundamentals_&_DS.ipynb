Question 1. What is mutability in Python?
   
   
   -> Mutability in Python

Introduction

Mutability is an important concept in Python that defines whether the value of an object can be changed after it has been created. Python is an object-oriented programming language where everything is treated as an object and stored in memory. Each object has an identity, a type, and a value. Based on whether the value of an object can be modified or not, Python objects are classified into mutable and immutable objects.

Understanding mutability is essential for proper memory management, efficient programming, and avoiding unexpected behavior in functions and programs.

Objects and Memory Representation in Python

In Python, variables do not store values directly. Instead, they store references to objects in memory.

Diagram 1: Variable Referencing an Object
a ───────► 10
          (object in memory)

Here:

a is a variable (reference)

10 is an object stored in memory

If the object changes without creating a new one, it is mutable. If a new object is created, it is immutable.

Mutable Objects in Python
Definition

Mutable objects are those objects whose values can be changed after creation without changing their memory location.

Characteristics of Mutable Objects

Modification is allowed after creation

Memory address remains the same

Efficient for frequent updates

Examples of Mutable Data Types

List

Dictionary

Set

Bytearray

Example: List (Mutable)
my_list = [1, 2, 3]
my_list.append(4)
Diagram 2: Mutable Object Modification
Before modification:
my_list ───► [1, 2, 3]

After modification:
my_list ───► [1, 2, 3, 4]



Example: Dictionary (Mutable)
student = {"name": "Amit", "age": 20}
student["age"] = 21
Diagram 3: Dictionary Modification
student ───► {"name": "Amit", "age": 21}

The dictionary is modified without creating a new object.

Immutable Objects in Python
Definition

Immutable objects are those objects whose values cannot be changed after creation. Any modification results in the creation of a new object in memory.

Characteristics of Immutable Objects

Cannot be modified

New object is created on change

Safer and more reliable

Examples of Immutable Data Types

Integer

Float

String

Tuple

Boolean

Example: Integer (Immutable)
x = 5
x = x + 10
Diagram 4: Immutable Object Modification
Before modification:
x ───► 5

After modification:
x ───► 15


Example: String (Immutable)
name = "Python"
name = name + " Language"
Diagram 5: String Immutability
Before:
name ───► "Python"

After:
name ───► "Python Language"

A new string object is created instead of modifying the original one.

Difference Between Mutable and Immutable Objects
Feature	Mutable Objects	Immutable Objects
Change allowed	Yes	No
Memory location	Same	New location
Data safety	Less	More
Examples	list, dict, set	int, str, tuple
Mutability and Function Behavior

When objects are passed to functions, mutability affects whether the original object is modified or not.

Example: Mutable Object in Function

    lst.append(100)

              append(100)

Inside function:

Advantages of Mutable Objects

Efficient for frequent data updates

Suitable for dynamic data structures

Saves memory in repeated modifications

Advantages of Immutable Objects

Prevent accidental modification

Better data security

Safe for multi-threaded programs

Suitable for constant values

Practical Importance of Mutability

Helps in choosing correct data structures

Important in function design

Prevents logical errors

Improves program efficiency

Essential for understanding Python memory model

Conclusion

Mutability in Python defines whether an object’s value can be modified after creation. Mutable objects like lists and dictionaries allow changes without creating new memory locations, whereas immutable objects like integers and strings create new objects when modified. Understanding mutability is crucial for efficient memory management, predictable function behavior, and safe programming practices. Therefore, mutability is a fundamental concept that plays a significant role in Python programming.


Question 2. Create a list and tuple in Python, and demonstrate how attempting to
change an element works differently for each.

  ->   Creating a List and Tuple in Python and Demonstrating Mutability Differences
Introduction

In Python, data structures are used to store collections of elements. Two commonly used data structures are lists and tuples. Although both are used to store multiple values in a single variable, they differ significantly in terms of mutability.

A list is a mutable data structure, meaning its elements can be modified after creation.
A tuple is an immutable data structure, meaning its elements cannot be changed once it is created.

This difference plays a crucial role in memory management, performance, and program behavior.

1. Creating a List in Python

A list is created using square brackets [ ].

Example:
my_list = [10, 20, 30, 40]
print("Original List:", my_list)
Output:
Original List: [10, 20, 30, 40]
Modifying an Element in a List

Since lists are mutable, we can change any element using indexing.

my_list[1] = 25
print("Modified List:", my_list)
Output:
Modified List: [10, 25, 30, 40]
Explanation:

The value at index 1 (which was 20) is changed to 25.

The list object remains the same in memory.

No error occurs.

Diagram 1: List Modification

Before modification:

my_list ───► [10, 20, 30, 40]
               0   1   2   3

After modification:

my_list ───► [10, 25, 30, 40]
               0   1   2   3

✔ Element changed
✔ Same object modified
✔ List is mutable

2. Creating a Tuple in Python

A tuple is created using parentheses ( ).

Example:
my_tuple = (10, 20, 30, 40)
print("Original Tuple:", my_tuple)
Output:
Original Tuple: (10, 20, 30, 40)
Attempting to Modify an Element in a Tuple

Since tuples are immutable, attempting to change an element will result in an error.

my_tuple[1] = 25
Output:
TypeError: 'tuple' object does not support item assignment
Explanation:

Tuples do not allow modification after creation.

Python raises a TypeError.

The original tuple remains unchanged.

Diagram 2: Tuple Modification Attempt

Before modification:

my_tuple ───► (10, 20, 30, 40)
                0   1   2   3

Attempt to change:

my_tuple[1] = 25
      
After attempt:

my_tuple ───► (10, 20, 30, 40)

✔ Element cannot be changed
✔ Error generated
✔ Tuple is immutable

3. Memory Behavior Comparison

We can use the id() function to check memory identity.

List Example:
my_list = [1, 2, 3]
print(id(my_list))
my_list[0] = 100
print(id(my_list))

The memory address remains the same.

Tuple Example:
my_tuple = (1, 2, 3)
print(id(my_tuple))
my_tuple = (100, 2, 3)
print(id(my_tuple))

A new object is created, so the memory address changes.

4. Key Differences Between List and Tuple
Feature	List	Tuple
Syntax	[ ]	( )
Mutability	Mutable	Immutable
Element modification	Allowed	Not allowed
Error on modification	No	Yes (TypeError)
Performance	Slightly slower	Faster
5. Practical Importance

Lists are used when data needs to be modified frequently.

Tuples are used when data should remain constant.

Tuples are safer and can be used as dictionary keys.

Lists consume slightly more memory compared to tuples.

Conclusion

Lists and tuples are important data structures in Python, but they differ mainly in terms of mutability. Lists are mutable, allowing modification of elements after creation, whereas tuples are immutable and do not allow any changes once defined. Attempting to modify a tuple results in a TypeError. Understanding this difference is essential for efficient memory usage, data safety, and correct program behavior. Therefore, choosing between a list and a tuple depends on whether the data needs to be modified or remain constant.

Question 3. What are logical operators in Python?
   ->  Logical Operators in Python
Introduction

Logical operators in Python are used to combine conditional statements and evaluate multiple conditions together. They return Boolean values (True or False) based on the logical relationship between expressions. Logical operators are commonly used in decision-making statements such as if, while, and conditional expressions.

In Python, logical operators are keywords rather than symbols. They help in controlling the flow of a program by evaluating multiple conditions efficiently.

Python provides three main logical operators:

and

or

not

1. The and Operator


The and operator returns True only if both conditions are True. If any one condition is False, the result is False.

Truth Table for and
Condition 1	Condition 2	Result
True	True	True
True	False	False
False	True	False
False	False	False
Example
a = 10
b = 5

print(a > 5 and b > 2)
Output:
True
Explanation:

a > 5 → True

b > 2 → True

Since both are True, result is True.

Diagram Representation of and
Condition A ----\
                 AND ----> Result
Condition B ----/

Both conditions must be True for final result to be True.

2. The or Operator
Definition

The or operator returns True if at least one condition is True. It returns False only when both conditions are False.

Truth Table for or
Condition 1	Condition 2	Result
True	True	True
True	False	True
False	True	True
False	False	False
Example
x = 3
y = 8

print(x > 5 or y > 5)
Output:
True
Explanation:

x > 5 → False

y > 5 → True

Since at least one is True, result is True.

Diagram Representation of or
Condition A ----\
                 OR ----> Result
Condition B ----/

If any one condition is True, the result becomes True.

3. The not Operator
Definition

The not operator is used to reverse the Boolean value of a condition. If the condition is True, it returns False; if False, it returns True.

Truth Table for not
Condition	Result
True	False
False	True
Example
a = 10
print(not(a > 5))
Output:
False
Explanation:

a > 5 → True

not True → False

Diagram Representation of not
Condition ----> NOT ----> Opposite Result
Short-Circuit Evaluation

Python uses short-circuit evaluation with logical operators.

In and, if the first condition is False, Python does not check the second condition.

In or, if the first condition is True, Python does not check the second condition.

Example:
print(False and True)   # Second condition not evaluated
print(True or False)    # Second condition not evaluated

This improves program efficiency.

Using Logical Operators in Conditional Statements

Logical operators are mostly used with if statements.

Example:
age = 20
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")

Logical operators allow checking multiple conditions in a single statement.

Difference Between Logical Operators
Operator	Meaning	Condition for True
and	Logical AND	Both conditions True
or	Logical OR	At least one True
not	Logical NOT	Reverses result
Practical Applications

Used in decision-making statements.

Used in loops to control execution.

Used in input validation.

Used in complex condition checking.

Essential in real-world applications like authentication systems.

Advantages of Logical Operators

Simplifies complex conditions.

Reduces number of nested if statements.

Improves readability of code.

Increases program efficiency due to short-circuit evaluation.

Conclusion

Logical operators in Python (and, or, and not) are essential tools for combining and manipulating conditional statements. They return Boolean values based on logical relationships between expressions. The and operator requires both conditions to be True, the or operator requires at least one True condition, and the not operator reverses the Boolean value. Logical operators are widely used in decision-making and control structures, making them fundamental to Python programming.

Question 5. Describe how dictionaries store data.


   ->  How Dictionaries Store Data in Python

Introduction

A dictionary in Python is a built-in data structure used to store data in the form of key–value pairs. Unlike lists and tuples, which store data in a sequential manner using indexes, dictionaries store data using unique keys. Dictionaries are highly efficient for retrieving data because they are implemented using a special internal mechanism called a hash table.

Understanding how dictionaries store data is essential for writing efficient Python programs and optimizing data retrieval operations.

1. Definition of Dictionary

A dictionary is a mutable, unordered collection of elements stored in key–value format.

Syntax:
my_dict = {
    "name": "Rahul",
    "age": 21,
    "city": "Delhi"
}

Here:

"name", "age", "city" are keys

"Rahul", 21, "Delhi" are values

2. How Dictionaries Store Data Internally

Python dictionaries store data using a hash table.

A hash table works by:

Taking a key.

Applying a hash function to generate a hash value.

Using that hash value to determine where to store the key–value pair in memory.

This allows dictionaries to achieve very fast data access (average time complexity: O(1)).

3. Internal Working of Dictionary Storage
Step 1: Key is Passed

When we insert:

student = {"name": "Amit"}

Python performs:

hash("name")
Step 2: Hash Value Generated

The hash function converts the key into a unique integer value.

Example (conceptual):

hash("name") → 4587321
Step 3: Index Calculation

The hash value is used to determine an index in memory where the key-value pair will be stored.

Diagram: Dictionary Storage Mechanism
Key: "name"
        │
        ▼
   Hash Function
        │
        ▼
   Hash Value (e.g., 4587321)
        │
        ▼
  Memory Index (Location)
        │
        ▼
Stored as:  "name" : "Amit"
4. Memory Representation of Dictionary

Example:

student = {
    "name": "Amit",
    "age": 21
}

Conceptual Memory Layout:

Memory Index      Key        Value
-------------------------------------
Index 1           "name"     "Amit"
Index 5           "age"      21

The data is not stored sequentially like lists.
It is stored based on the hash of the key.

5. Key Characteristics of Dictionary Storage
1. Keys Must Be Unique

Each key in a dictionary must be unique.

data = {"a": 1, "a": 2}

Output:

{'a': 2}

The second value overwrites the first.

2. Keys Must Be Immutable

Keys must be of immutable types such as:

int

float

str

tuple

Mutable types like lists cannot be used as keys because their hash value could change.

Example (Invalid):

my_dict = {[1, 2]: "value"}  # Error
3. Fast Lookup Time

Dictionaries provide:

Insertion → O(1)

Deletion → O(1)

Access → O(1)

This makes dictionaries much faster than lists for searching data.

6. Handling Hash Collisions

Sometimes two different keys may produce the same hash value. This is called a hash collision.

Python handles collisions internally using techniques such as:

Open addressing

Probing methods

Conceptual Diagram of Collision:

Key1 → Hash → Index 3
Key2 → Hash → Index 3  (Collision)

Python resolves internally and stores safely.

This ensures correct retrieval of data.

7. Example Demonstrating Storage and Retrieval
student = {
    "name": "Amit",
    "age": 21
}

print(student["name"])

When "name" is accessed:

Hash of "name" is calculated.

Memory index is found.

Value "Amit" is returned.

This makes retrieval extremely efficient.

8. Difference Between Dictionary and List Storage
Feature	Dictionary	List
Storage Method	Hash table	Sequential
Access Method	Key-based	Index-based
Lookup Speed	O(1)	O(n)
Key Requirement	Unique & Immutable	Not required
9. Practical Applications

Dictionaries are widely used in:

Database records

JSON data handling

Configuration storage

Counting frequency of elements

Caching systems

Example:

marks = {
    "Math": 90,
    "Science": 85
}

Here subject names are keys, and marks are values.

10. Advantages of Dictionary Storage

Fast data retrieval.

Efficient memory management.

Easy mapping between related data.

Flexible data storage.

Supports dynamic data updates.

Conclusion

Dictionaries in Python store data using a hash table mechanism where each key is processed through a hash function to determine its storage location in memory. This method enables extremely fast data retrieval and efficient storage. Keys must be unique and immutable to maintain consistency in hashing. Unlike lists that store data sequentially, dictionaries store data based on hash values, making them highly suitable for applications requiring quick lookups and structured data representation. Therefore, dictionaries are one of the most powerful and efficient data structures in Python.

 Question 9.  Why might you use a set instead of a list in Python?
              ->  1. Introduction

In Python, both lists and sets are built-in data structures used to store collections of items. However, they serve different purposes and have different properties.

A list is an ordered, mutable collection that allows duplicate elements.
A set is an unordered, mutable collection that does not allow duplicate elements.

Choosing between a set and a list depends on the requirements of the program.

2. Definition of List and Set
List

A list stores elements in a specific order and allows duplicate values.

Example:
numbers = [1, 2, 2, 3, 4]
print(numbers)

Output:

[1, 2, 2, 3, 4]

Duplicates are allowed.

Set

A set stores unique elements and does not maintain order.

Example:
numbers = {1, 2, 2, 3, 4}
print(numbers)

Output:

{1, 2, 3, 4}

Duplicates are automatically removed.

3. Reasons to Use a Set Instead of a List
1. To Remove Duplicate Elements

One major reason to use a set is to eliminate duplicates automatically.

Example:
marks = [85, 90, 85, 70, 90]
unique_marks = set(marks)
print(unique_marks)

Output:

{85, 90, 70}

Sets automatically remove repeated values.

2. Faster Membership Testing

Sets are much faster than lists when checking if an element exists.

Example:
numbers = {10, 20, 30, 40}

if 20 in numbers:
    print("Found")

Membership checking in:

List → Slower (O(n))

Set → Faster (O(1) average time complexity)

This makes sets ideal for large datasets.

3. Mathematical Set Operations

Sets support mathematical operations like:

Union (|)

Intersection (&)

Difference (-)

Symmetric Difference (^)

Example:
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)  # Union
print(A & B)  # Intersection

Output:

{1, 2, 3, 4, 5}
{3}

Lists do not directly support these operations.

4. Preventing Duplicate Data

When storing:

Unique usernames

Unique email IDs

Unique product IDs

Sets ensure data uniqueness automatically.

5. Efficient Data Comparison

Sets are useful when comparing two collections of data.

Example: Finding common students in two classes.

classA = {"Aman", "Ravi", "Sita"}
classB = {"Ravi", "Sita", "Neha"}

print(classA & classB)

Output:

{'Ravi', 'Sita'}
4. Key Differences Between List and Set
Feature	List	Set
Order	Ordered	Unordered
Duplicates	Allowed	Not allowed
Indexing	Supported	Not supported
Membership Test Speed	Slower	Faster
Mathematical Operations	Not built-in	Supported
5. When to Use a Set

You should use a set when:

You need unique elements only.

Order of elements does not matter.

You need fast membership testing.

You want to perform set operations.

You are removing duplicates from data.

6. When Not to Use a Set

Avoid using a set when:

Order matters.

You need indexing (like list[0]).

You want duplicate values.

You need slicing.

7. Practical Real-Life Examples

Storing unique website visitors

Finding common friends in social media

Removing duplicate entries from a database

Checking forbidden words in a text

Comparing large datasets

8. Advantages of Using Set

Automatically removes duplicates

Faster searching

Supports mathematical operations

Efficient for large datasets

Cleaner code for uniqueness tasks

9. Conclusion

A set should be used instead of a list when uniqueness, speed, and mathematical operations are important. While lists are useful when order and duplicates are required, sets are more efficient for handling unique data and performing quick membership checks. Therefore, choosing between a set and a list depends on the specific problem requirements.
